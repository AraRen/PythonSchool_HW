```python
# 使用10*10均勻分配資料
np.random.seed(0)
uniform_data = np.random.rand(10,10)
sns.heatmap(uniform_data)
```

```python
sns.heatmap(uniform_data,vmin=0.5,vmax=0.8)
```

```python
sns.heatmap(uniform_data,center=0.5)
```

```python
#抽取標準常態分配10*10數據，以零為中心點
np.random.seed(0)
normal_data = np.random.randn(10,10)
sns.heatmap(normal_data,center=0)
```

```python
#使用均勻分配
sns.heatmap(uniform_data,cmap='Greens')
```

```python
#使用標準常態分配
sns.heatmap(normal_data,cmap='Blues')
```

```python
#設定標題
sns.heatmap(normal_data)
plt.title('heatmap')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#設定刻度
sns.heatmap(normal_data,xticklabels=2,yticklabels=3)
```

```python
#設定隨機種子
np.random.seed(0)

#存均勻分配[0~1]之間抽出3*3陣列
uniform_data = np.random.rand(3,3)
sns.heatmap(uniform_data)
```

```python
#設為正方形
sns.heatmap(uniform_data,square=True)
```

```python
#載入飛行數據並檢視前五筆資料
flights = sns.load_dataset('flights')
flights.head()
```

```python
#飛行數據統計資料
flights.describe()
```

```python
#使用pivot函數將資料轉為DataFrame結構
flights_df = flights.pivot('month','year','passengers')
flights_df
```

```python
#繪製飛行數據熱力圖
sns.heatmap(flights_df)
```
